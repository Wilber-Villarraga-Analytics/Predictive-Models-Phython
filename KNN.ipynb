# K VECINOS MÁS CERCANOS (KNN) 

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
# Datos
baseball = pd.read_csv("train.csv")
baseball = baseball[["plate_x",
                     "plate_z",
                     "launch_angle",
                     "launch_speed",
                     "is_home_run"]]
```

```python
# Separar  la variable objetivo
X = baseball.drop(columns='is_home_run')
y = baseball['is_home_run'].values

# Separar en entrenamiento y prueba
X_train,X_test,y_train,y_test = train_test_split(X,y,
                                                 test_size = 0.2,
                                                 random_state = 2352,
                                                 stratify = y)
```

```python
# Definicion del modelo
knn_model = KNeighborsClassifier(n_neighbors = 10,
                                 weights = 'uniform',
                                 p = 1) # distancia de manjatan
```

```python
# Preprocesamiento
scaler = StandardScaler()
imputer = SimpleImputer(strategy = "mean") # la media de la distancias
```

```python
# Pipeline
pipe = Pipeline(steps = [
    ("Imputacion", imputer),
    ("Scaler", scaler),
    ("knn",knn_model)    
    ])
```

```python
# Entrenamiento
pipe.fit(X_train,y_train)

# Prediccion en test
pipe.predict(X_test)

# Accuracy en test
pipe.score(X_test,y_test)
```

```python
# Modelo
knn = KNeighborsClassifier()

# Pipeline
pipe = Pipeline(steps = [
    ("Imputacion", imputer),
    ("Scaler", scaler),
    ("knn",knn)    
    ])
```

```python
# Espacio a buscar
param_grid = {"knn__n_neighbors": np.arange(1,15),
              "knn__weights": ["uniform","distance"],
              "knn__p": [1,2]}
```

```python
# Optimizacion
knn_search = GridSearchCV(pipe,
             param_grid,
             cv = 5)
```

```python
knn_search.fit(X_train,y_train)
```

```python
#  Resultados
knn_search.best_score_
```

```python
knn_search.best_params_
```

# Conclucion

A partir del proceso de búsqueda de hiperparámetros, se determinó que el modelo K-Nearest Neighbors (KNN) presenta su mejor desempeño utilizando 13 vecinos (n_neighbors = 13), con distancia Manhattan (p = 1) y ponderación uniforme (weights = 'uniform').

La elección de un valor relativamente alto de k permite reducir la sensibilidad al ruido y a valores atípicos, logrando una frontera de decisión más estable. El uso de la distancia Manhattan resulta adecuado cuando las variables presentan escalas comparables y relaciones lineales, mientras que la ponderación uniforme indica que todos los vecinos influyen de igual manera en la clasificación final.

El tiempo de entrenamiento elevado observado durante el ajuste del modelo es esperado, ya que KNN es un algoritmo basado en instancias que requiere el cálculo de distancias para múltiples combinaciones de parámetros durante la validación cruzada. A pesar de este costo computacional, el modelo ofrece un desempeño consistente y confiable para el conjunto de datos analizado.

En conclusión, el modelo KNN ajustado con estos hiperparámetros constituye una alternativa adecuada para el problema de clasificación planteado, siempre que se disponga de recursos computacionales suficientes y un tamaño de datos manejable.
